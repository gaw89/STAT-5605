##### The Gentle Lentil - Case Analysis

### Introduction
In this case, we are presented with the story of Sanjay Thomas, a second-year MBA student at MIT's Sloan School of Management.  He is in the midst of a dilemma of having to choose between a job offer with a top management consulting firm and the possibility of opening his own restaurant - something he has always wanted to do.  After much consideration, Sanjay determines that the lifestyle tradeoffs between the two options are comparable, and the decision will ultimately come down to whichever provides the better income.

The offer from the management consulting firm is straightforward, `$`80,000 for the first year with the firm.  Sanjay finds that the earnings potential of opening his own restaurant, however, is much more complicated.  Sanjay has put a lot of thought and study into the revenue potential of a restaurant like the one he is considering, but due to the many factors involved, it is impossible to simply calculate the expected revenue/profit.  To get a good understanding of the plausible income from this venture, Sanjay must turn to simulation methods so that he can combine the various distributions of cost and revenue to determine a feasible range for the income of his restaurant concept.  To complicate things further, Sanjay has received an offer from his aunt which would make allow him to have a guaranteed monthly income of `$`3,500 per month but would severely limit any potential income over `$`9,000 (she would take 90% of all profit over `$`9,000).

Let's dive into the questions from the case.

### Case Assignment

#### (a) Without considering the partnership opportunity, what would be Sanjay's expected monthly salaray at Gentle Lentil? How does this compare with his monthly salary at the consulting firm?

```python
import numpy as np
from scipy import stats
from ggplot import ggplot, aes, ggtitle, geom_density
import pandas as pd
np.random.seed(123)

def num_meals_sold(N=1):
    return np.random.normal(3000, 1000, N)

def labor_cost(N=1):
    return np.random.uniform(5040, 6860, N)

def get_meal_price(N):
    probs = [0, .25, .6, .9, 1]
    prices = [20., 18.5, 16.5, 15.]

    samples = np.random.random(N)

    price_per_month = []
    
    for prob in samples:
        for i in range(0, len(probs) - 1):
            if probs[i] <= prob < probs[i+1]:
                price = prices[i]
                price_per_month.append(price)
                
    return np.array(price_per_month)

def profit(meals_sold, labor_cost, meal_price):
    
    variable_cost_per_meal = 11
    fixed_cost = 3995
    
    # Total revenue for the month
    revenue = meals_sold * meal_price
    
    total_variable_cost = variable_cost_per_meal * meals_sold
    
    # Total cost of operations for the month
    cost = labor_cost + fixed_cost + total_variable_cost
    
    profit = revenue - cost
    
    return profit

n_months = 10000

meals_sold = num_meals_sold(n_months)
labor = labor_cost(n_months)
meal_price = get_meal_price(n_months)

profits = profit(meals_sold, labor, meal_price)

# is this normal?
print("Kurtosis of a normal distribution should be 0.  Kurtosis of income: %.2f" % stats.kurtosis(profits))
# close enough

mean = profits.mean()
std = profits.std()

print("Average monthly income from Gentle Lentil (10,000 simulated months): $%.2f" % mean)
print("Standard deviation of monthly income from Gentle Lentil (10,000 simulated months): $%.2f" % std)

print("Monthly income from consulting firm: $%.2f" % (80000 / 12))

df = pd.DataFrame(profits)
plot = ggplot(aes(x="profits"), data=df) + geom_density() + ggtitle("Density Plot of Monthly Profit")
print(plot)
```

As we see above, profit is an approximately normal distribution (close enough that we can reasonably use properties of the normal distribution) with a mean of ~`$`11,000 and a standard deviation of `$`8,700.  Since this variable is close to normal, we can use the cumulative probabilities from the normal distribution to understand the probability of Sanjay having a profit below `$`5,000 in a month (Sanjay's minimum required monthly income).

```python
# Z-score of $5,000
z_score = (5000 - mean) / std

print("Probability of profit <= $5,000: %.3f" % stats.norm.cdf(z_score))
```

We see here that the probability of Sanjay having a monthly salary less than or equal to `$`5,000 is approximately .25.  This may be an unacceptable level of risk for Sanjay, especially when the guaranteed monthly salary from the consulting firm is almost `$`7,000.
